 <font size = 3>Make the severity variable(column) categorical (qualitative/discrete) and further from      nominal to dichotomous : 0 -> alive, 1 -> dead</font>

```python
# implement routine to make variable into  categorical dichotomous

def mke_var_cat(x):
    if int(x) == 1 or int(x) == 3 or int(x) == 4:
        x = 0
    if int(x) == 2:
        x = 1
    return x
```

```python
# use apply routine to make another variable(column) categoricacl(qualitative/discrete)(dichotomous) with identifier deadoralive

df_accidents["deadoralive"] = df_accidents["Severity"].apply(mke_var_cat).copy()
```

<font size = 3>The majority class of no deaths has 3 million while the minority class of deaths has only a mere 82 thousand. Any machine learning model will have a high accuracy due to its predicting the majority class and completely ignoring the minority class. Since we have a rather bigger data with 3 million rows i chose to down sample the majority class rather then upsample the minority class. </font>

```python
# Separate majority and minority classes
df_majority = df_accidents[df_accidents.deadoralive==0]
df_minority = df_accidents[df_accidents.deadoralive==1]
 
# Downsample majority class
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=82151,     # to match minority class
                                 random_state=123) # reproducible results
 
# Combine minority class with downsampled majority class
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
 
# Display new class counts
df_downsampled.deadoralive.value_counts()
```

<font size = 3>list containing validation set & test set  accuracy scores for the classification ML models with identifier acc_scr_classification_ML_models</font>

<font size = 3> Naive Bayes Classifier</font>

```python
# construct pipelines for different hyperparameter tuning versions of said model:

start_time_NBC = time.time()

pipe_NBC_1 = Pipeline([("bnb", BernoulliNB())])

pipe_NBC_2 = Pipeline([("bnb", BernoulliNB(alpha = 0))])

# list of pipelines for iteration

pipelines = [pipe_NBC_1, pipe_NBC_2]

# dictionary of pipelines

pipe_dict = {0 : "pipe_NBC_1", 1: "pipe_NBC_2"}

# fit the pipelines to said model

for pipe in pipelines:
    pipe.fit(X_train1, y_train1)

# compare cross_val_score mean of different NBC pipelines on validation dataset

for idx, val in enumerate(pipelines):
    print('%s hyperparameter tuning version cross_val_score mean: %.3f' % (pipe_dict[idx], cross_val_score(val, X_validate, y_validate).mean()))

# identify the best mean cross_val_score  NBC pipeline on validation dataset

best_cross_val_score_mean = 0.0
best_NBC_pipe_hyperparameter_tuning_version = 0
best_NBC_pipe = ''
for idx, val in enumerate(pipelines):
    if cross_val_score(val, X_validate, y_validate).mean() > best_cross_val_score_mean:
        best_cross_val_score_mean = cross_val_score(val, X_validate, y_validate).mean()
        best_NBC_pipe_hyperparmeter_tuning_version = idx
        best_NBC_pipe = val
print(' NBC hyperparameter tuning version with best mean cross_val_score: %s' % pipe_dict[best_NBC_pipe_hyperparmeter_tuning_version])

# obtain the  accuracy score for NBC pipeline with best mean cross_val_score in the validation set

lst_acc_scr_classification_ML_models.append((accuracy_score(y_validate, pipelines[best_NBC_pipe_hyperparmeter_tuning_version].predict(X_validate))))

print("The accuracy score for pipeline with best mean cross_val_score in the validation set is %.3f" %(accuracy_score(y_validate, pipelines[best_NBC_pipe_hyperparmeter_tuning_version].predict(X_validate))))

# obtain the accuracy score for  NBC pipeline with best mean cross_val_score in the test set

lst_acc_scr_classification_ML_models.append((accuracy_score(y_test, pipelines[best_NBC_pipe_hyperparmeter_tuning_version].predict(X_test))))

print("The accuracy score for pipeline with best mean cross_val_score in the test set is %.3f" %(accuracy_score(y_test, pipelines[best_NBC_pipe_hyperparmeter_tuning_version].predict(X_test))))

print("--- %s seconds ---" % (time.time() - start_time_NBC))
```
